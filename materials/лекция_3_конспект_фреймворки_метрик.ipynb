## Фреймворки для продуктовых метрик

Мы познакомились с большим количеством известных метрик, теперь встают важные вопросы - с чего начать и как это все организовать.

Для проектирования метрик существует несколько фреймворков (подходов), которые мы рассмотрим. Они могут помочь, например, если вы запустили продукт и хотите начать отслеживать основные показатели, или у вас уже есть построенные метрики и дашборды, и вы хотите организовать их в понятные структуры, разделить зоны ответсвенности или нарисовать связи между метриками.



### 1. Heart Framework

Первый фреймворк который мы рассмотрим - **Heart Framework** от компании Google. Фреймворк разбит на пять категорий: счастье, вовлеченность, принятие, удержание и успех выполнения задачи. Все вместе — это набор ux-метрик, которые позволяют получать обратную связь от пользователей и реагировать на неё в реальном времени.

Для работы с фреймворком Google предлагает систему «цели – сигналы – метрики». Для каждой категории HEART определяют:

**Цель** — что нужно получить; Цели - это общие задачи. В случае категории "Hapiness" целью может быть “Пользователи находят приложение полезным”.

**Сигнал** — что в поведении пользователей укажет на успех или неудачу в достижении цели; Сигналом может быть “Пользователи рекомендуют приложение своим друзьям”.

**Метрику** — количественное измерение сигнала. Измерить то как пользователи рекомендуют приложение своим друзьям, может, например, метрика K-factor, которую мы обсуждали на занятии.

## 2. MixPanel Framework

В продуктовой аналитике существует такое понятие как **North Star Metric**. NSM - это показатель, который лучше всего отражает ценность продукта для пользователя, подробнее можно прочитать тут - https://gopractice.ru/product/the-north-star-metric-guide/ 




Компания MixPanel предлагает фреймворк для организации работы и раскладывания метрик, который основан на нескольких основных идеях:

1. У каждого продукта существует фокусная метрика (Focus Metric), ключевая метрика, которая определяет стратегию продукта. Это аналог NSM, но с менее жесткой формулировкой. В качестве Focus Metric необязательно выделять только одну единственную метрику, но в то же время их не должно быть слишком много. В любой период времени у вашей компании должна быть стратегия, которая определяет что мы хотим растить в данный момент, распылять фокус на множество показателей может быть неэффективно.
2.  Фокусные метрики - обычно очень высокоуровневые. Например, это может быть DAU или GMV. Существует большое количество точек влияния на них, то есть их можно разложить на составляющие. Например, DAU можно растить привлекая новых пользователей или увеличивая удержание и Sticky Factor. Чтобы привлекать новых пользователей можно использовать разные маркетинговые стратегии, отдел продаж, или работать через существующую аудиторию - стимулировать их рекомендовать ваш продукт друзьям. Каждый из этих процессов можно разложить на метрики более низкого уровня - показатели маркетинга / продаж /рекомендаций. Соответсвенно идея предложенного фреймворка - это разбить метрики влияющие на целевые показатели по смыслу и по уровню
3.  По вертикали фреймворк предлагает выделить пять смысловых категорий, которые завязаны на цикл пользователя в продукте
    - Reach или охват  — это общее количество людей, которые использовали продукт за последний период времени.
    - Activation  — это основополагающий шаг, который помогает новому пользователю стать активным пользователем. Как известно,
Facebook определил добавление 7 друзей за 10 дней в качестве пользовательской активации, когда они были стартапом. Они
обнаружили, что это ключевая веха в обеспечении долгосрочного использования, и поэтому сделали добавление друзей центральной частью
их опыт адаптации. Для других типов продуктов активацию можно определить как регистрацию, совершение первой покупки, просмотр
первых пяти видео или внесение двух депозитов в течение определенного периода времени.
    - Active Users - Активные пользователи — это люди, которые совершили ключевое действие и получили пользу от вашего продукта внутри компании.
    - Engagement  -  отличается от активного использования, поскольку  измеряет более глубокий уровень вовлечения в
продукт. Engagement метрики отвечают на вопрос «Насколько вовлечены ваши активные пользователи?»
    - Retention — сюда собтсвенно относятся метрики удержания, это может быть 7-day Retention, 30-days Retention, Churn Rate и тд
    - Business specific - метрики характерные для вашего бизнеса. Например, сайт знакомств,может использовать показатель «хороший отток» для обозначения пользователей, которые находят длительные отношения и покидают приложение. Хотя это влечет за собой потерю пользователей, это хорошо для бизнеса, потому что довольные клиенты порекомендуют друзьям и сами вернутся, если когда-нибудь снова окажутся в поиске.

5.  По горизонтали фреймворк выделяет два уровня метрик: Level 1 Metrics - показатели, которые либо непосредственно влияют на фокусную метрику, либо служат индикаторами здорового роста продукта. И Level 2 Metrics - метрики, которые раскладывают метрики первого уровня на составляющие. Это может быть например, разложение метрики первого уровня на различные разрезы (платформы/страны) и тд

## 3. AARRR Framework

**AARRR Funnel** или пиратская воронка - одна из самых популярных моделей для стартапов и торговых компаний для измерения роста и успеха. Она была представлена Дэйвом Макклюром, основателем Practical Venture Capital, и также называется Pirate Metrics или Пиратский фреймворк. Он разбивает жизненный цикл вашего пользователя на 5 этапов и сопоставляет их с соответствующими показателями – начиная с момента первого контакта и заканчивая моментом, когда потенциальный клиент приносит вам деньги.
 
Один из распространенных ответов на вопрос «Как растет ваш продукт?» это изображение воронки. 

Эти этапы также можно использовать, чтобы разбить ваши метрики по смыслу и по зонам ответсвенности.

## 5. Иерархия метрик

Подходы из пунктов 2, 3, 4 могут помочь вам в организации отчетности и дашбордов и распределении зон ответсвенности между командами, но есть еще один важный инструмент для отображения метрик продукта - это иерархия метрик. Идея иерархии метрик заключается в том, что вы снова выделяете фокусную метрику для продукта/фичи или команды, но теперь расскладываете ее не по смысловым категориям, а именно последовательно разбиваете на ее составляющие. Каждое следующие разбиение - это именно комбинация которая формирует предыдущую метрику.


**Пример для сервиса видеоконференций:**

Ваша фокусная метрика - это общий WAU. 
- WAU для вас - это количество уникальных пользователей в неделю, которые приняли участие хотя бы в одной конференции. Это WAU можно разделить на создателей (хостов) и участников конференции. Вашим стимулом роста являются именно создатели конференций, поэтому дальше фокусируемся на раскладке WAU хостов.
- WAU хостов состоит из новых пользователей (new), тех что были и на прошлой неделе и на этой (retained), тех что были более недели назад и вернулись на этой неделе (reactivated), и иногда отдельно выделяют тех кто не был уже очень давно и вернулся (например - не был в сервисе три месяца) (reanimated)
- Количество новых хостов в свою очередь можно разложить на каналы привлечения. Создатель конференции может быть привлечен маркетингом. (WAU marketing). Дальше эту метрику можно было бы разложить на маркетинговые каналы (Например те кто пришел из поисковых запросов, тех кто пришел с рекламы и тд, а также конверсии. Например количество хостов привлеченных с рекламы = количество посетителей сайта с рекламы * конверсию посетителей сайта с рекамы в создание конференции. Количество посетителей сайта с рекламы = количество людей увидевших рекламу * конверсию перехода на ваш сайт. И так далее
- Количество удержанных хостов (retained) раскладывается на количество хостов за предыдущий период * Retention Rate
- Retention Rate можно дальше разложить на метрики качества и метрики пользовательского вовлечения или удовлетворенности и так далее

В реальности иерархия может быть очень большой, но это крайне полезный инструмент, который полезно проработать. Она помогает нам определить точки влияния и точки роста, и генерить идеи с помощью чего мы можем влиять на каждую метрику. Также если вы видете рост или падение какой-то метрики визуализация иерархии может вас направить какие причины роста или подения нужно идти проверять.

Такой подход также может помочь на собеседованиях - часто на собесах ПА бывают кейсы по типу "Вам нужно повысить метрику N, что будете делать, какие эксперименты проводить" или "Упала метрика X как будете разбираться"

**Пример иерархии для команды, отвечающей за спрос на маркетплейсе такси (приложени такси)**

Ключевой метрикой для такой команды может быть суммарное количество часов спроса на поездки = Demand Hours. (Тут взяли не количество клиентов делающих заказ или количество заказов, так как заказ для одного клиента может быть выполнен за 10 минут, а может быть за 2 часа. У нас есть емкость водителей в виде человеко-часов которые они могут работать, поэтому и спрос будем оценивать в часах).

- Demand Hours раскладывается на demand hours per user и WAU
- Demand Hours Per User можно разложить на частоту заказов юзера в неделю и среднюю дистанцию заказа
- WAU раскладываем на новых / удержанных с предыдущей недели и реактивированных
- Новых можем разложить на каналы заказа (Первый заказ в приложении / первый заказ на сайте)
- Первый заказ на сайте можем разложить на WAU сайта и конверсию из посещения в заказ
- Удержанных можем разложить на юзеров за прошлую и ретеншн (так как удержанные = юзеры прошлой неделе * ретеншн)
- Реактивированных могли бы дальше разложить на реактивированных органически (вернулись сами) и маркетинговые активности по реактивации (пуши скидки и тд)
